```python
documents = [
    "Artificial intelligence (AI) is the field of computer science focused on creating systems that can perform tasks typically requiring human intelligence.",
    "It includes areas such as machine learning, natural language processing, computer vision, and robotics.",
    "AI systems learn from data to recognize patterns, make decisions, and improve their performance over time.",
    "Today, AI is widely used in applications like virtual assistants, recommendation systems, healthcare, finance, and autonomous vehicles."
]
```

```python
import string

def retrieve(query, documents):
    cleaned_query = query.lower().translate(
        str.maketrans("", "", string.punctuation)
    )

    query_words = cleaned_query.split()
    results = []

    for doc in documents:
        for word in query_words:
            if word in doc.lower():
                results.append(doc)
                break

    return results[:2]
```

```python
def generate_response(query, retrieved_docs, chat_history):
    response = "Based on what I found:\n"

    for doc in retrieved_docs:
        response += f"- {doc}\n"

    response += "\nLet me know if you want more details."
    return response
```

```python
chat_history = []

print("Dialogue RAG System (type 'exit' to stop)\n")

while True:
    user_query = input("User: ")

    if user_query.lower() == "exit":
        print("Ending conversation.")
        break

    chat_history.append(user_query)

    retrieved_docs = retrieve(user_query, documents)
    response = generate_response(user_query, retrieved_docs, chat_history)

    print("\nAssistant:", response, "\n")
```
